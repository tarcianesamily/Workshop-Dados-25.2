```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


temperaturas = np.random.uniform(20, 35, 7).round(2) # vai gerar temperaturas aleatórias nesse range e com 2 casas decimais
dias_semana = ["Segunda", "Terça", "Quarta", "Quinta", "Sexta", "Sábado", "Domingo"]

# criando um dataframe
df = pd.DataFrame({'Dia':dias_semana, 'Temperatura °C':temperaturas})
print(df)

# calculando a média
media = np.mean(temperaturas)
print("A média de temperatura da semana é de {:.2f} °C.".format(media))

# gráfico
x = dias_semana
y = temperaturas

plt.plot(x, y, label='Variação da temperatura na semana') # legenda
plt.xlabel('Dias da semana')
plt.ylabel('Temperatura em °C')
plt.legend()
plt.show()
```
